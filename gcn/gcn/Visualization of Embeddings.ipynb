```python
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.manifold import SpectralEmbedding

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from scipy.special import softmax
```

```python
X = np.load('embed2.npy')
X = softmax(X, axis=1)
ally = np.load('data/ind.cora.ally')
testy = np.load('data/ind.cora.ty')
y = np.vstack([ally, testy])
Y = np.argmax(y, axis=1)
```

```python
Xsne = TSNE(n_components=2).fit_transform(X)
```

```python
plt.figure(figsize=[10,6])
plt.scatter(Xsne[:,0], Xsne[:,1], s=5, c=Y, cmap='gist_rainbow')
```

```python
plt.figure(figsize=[10,6])
Xpca = PCA(2).fit_transform(X)
plt.scatter(Xpca[:,0], Xpca[:,1], c=Y, cmap='gist_rainbow')
```
