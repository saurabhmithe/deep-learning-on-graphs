```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from scipy.special import softmax
```

```python
X = np.load('cora_embed.npy')
X = softmax(X, axis=0)
ally = np.load('data/ind.cora.ally')
testy = np.load('data/ind.cora.ty')
y = np.vstack([ally, testy])
Y = np.argmax(y, axis=1)
```

```python
Xsne = TSNE(n_components=2).fit_transform(X)
```

```python
plt.scatter(Xsne[:,0], Xsne[:,1], c=Y)
```
